```python
import numpy as np 
import pandas as pd 
import geopandas as gpd 
import contextily as ctx
from shapely.wkt import loads
import matplotlib.pyplot as plt 
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
edges_df = pd.read_csv('projects/tokyo_residential_above/network_inputs/edges_residual_demand.csv')
edges_df.iloc[0]
```

```python
edges_gdf = gpd.GeoDataFrame(edges_df, crs='epsg:4326', geometry=edges_df['geometry'].map(loads))
```

```python
ax0 = edges_gdf.iloc[0:10].plot(figsize=(10,10))
```

```python
edges_web_gdf = edges_gdf.iloc[0:100].to_crs(epsg=3857)

fig, ax = plt.subplots(1,1, figsize=(10,10))
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad=0.1)
edges_web_gdf.plot(column='length', lw=0.5, ax=ax, cax=cax, cmap='rainbow', legend=True, vmin=0, vmax=10)
ctx.add_basemap(ax, url=ctx.providers.Stamen.TonerLite, alpha=0.2)
fig.patch.set_facecolor('white')
fig.patch.set_alpha(0.7)
ax.set_title('Test', font={'size': 30})
plt.savefig('projects/tokyo_residential_above/visualization_outputs/test.png', transparent=False)
```
